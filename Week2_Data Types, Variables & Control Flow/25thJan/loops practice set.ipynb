```python
# Take inputs from user to make a list. Again take one input from user and search it in the list and delete that element, if 
# found. Iterate over list using for loop.
x=int(input('enter the no. of elements to be in list: '))
list_1=[0]
for i in list_1:
    if len(list_1)<x+1:
        y=int(input("enter the element: "))
        list_1.append(y)
print('your list is ready')
print(list_1)
z=int(input('enter the element to be searched:'))
for j in list_1:
    if j == z:
        print('item found...')
        list_1.remove(z)
        print(z,'is deleted.')
    else:
        print("item not found!!!")
print(list_1)        
```

```python
#Using range(1,101), make two list, one containing all even numbers and other containing all odd numbers. 
x=range(1,25)
list_even=[]
list_odd=[]
for i in x:
    if i%2==0:
        list_even.append(i)
    else:
        list_odd.append(i)
print("list for even nos.-",list_even)
print("list for odd nos.-",list_odd)
```

```python
#From a list containing ints, strings and floats, make three lists to store them separately.
x=['sahil',17,72.6,'rajeev',23,75.98,'sourav',39,68.7,'aman',680,'ronak',564,4.6,3.4]
l_int=[]
l_float=[]
l_str=[]
for i in x:
    if type(i)==int:
        l_int.append(i)
    if type(i)==float:
        l_float.append(i)
    if type(i)==str:
        l_str.append(i)
print('integers are:',l_int) 
print('float values are:',l_float)
print('strings are:',l_str)
```

```python
while True:
    # Prompt the user to enter a number
    num = int(input("Enter a number (0 to exit): "))

    # Check if the number is zero
    if num == 0:
        break   # Exit the loop if the number is zero

    # Otherwise, print the square of the number
    print(f"The square of {num} is {num**2}")
```

```python
total = 0

while True:
    num = int(input("Enter a number: "))
    if num < 0:
        break
    total += num

print("The sum is:", total)
```

```python
total1 = 0
total2 = 0

while True:
    num1 = int(input("Enter a number: "))
    num2 = int(input("Enter a number: "))
    print(num1+num2)
    if (num1 < 0) or (num2<0):
        break
    total1 += num1
    total2 += num2

print("The sum is:", total1)
print("The sum is:", total2)

```
