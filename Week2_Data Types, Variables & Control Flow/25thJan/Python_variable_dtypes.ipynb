```python
# some nested dictionary examples to understand more
student = {
    "name": "Alice",
    "age": 22,
    "courses": {
        "math": {"grade": "A", "credits": 3},
        "science": {"grade": "B", "credits": 4}
    }
}
student
```

```python
blog_tags = {"python", "coding", "tutorial", "python"}  # 'python' is duplicated
blog_tags.add("programming")  # Adds an item to the set
print(blog_tags)  # Output: {'python', 'coding', 'tutorial', 'programming'}
blog_tags.remove("coding")  # Removes an item from the set; raises KeyError if not found
print(blog_tags)  # Output: {'python', 'tutorial', 'programming'}
blog_tags.clear()  # Removes all items from the set
print(blog_tags)  # Output: set()
```

```python
# 2 sets operations examples
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
# Union
set_union = set_a | set_b # Combines all unique elements from both sets
print(set_union)  # Output: {1, 2, 3, 4, 5, 6}
# Intersection
set_intersection = set_a & set_b # Combines elements present in both sets
print(set_intersection)  # Output: {3, 4}
# Difference
set_difference = set_a - set_b # Elements in set_a but not in set_b
print(set_difference)  # Output: {1, 2}
# Symmetric Difference
set_sym_diff = set_a ^ set_b
print(set_sym_diff)  # Output: {1, 2, 5, 6}
```

```python
# f string vs .forrmat()
name = "Bob"
# Using f-string
f_string_greeting = f"Hello, {name}!"
print(f_string_greeting)

# Using .format()
format_greeting = "Hello, {}!".format(name)
print(format_greeting)
```

```python
# introduce Error handling in python with examples
# Error handling in Python is done using the try, except, else, and finally blocks. It allows you to gracefully handle exceptions (errors) that may occur during the execution of your code, preventing the program from crashing.
```

```python
# Example of error handling:
try:
    num1 = int(input("Enter a number: "))
    num2 = int(input("Enter another number: "))
    result = num1 / num2
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except ValueError:
    print("Error: Invalid input. Please enter a valid number.")
else:
    print("Result:", result)
```

```python
try:
    fruits = ["apple", "banana", "cherry"]
    index = int(input("Enter an index to access the fruit: "))
    print("Fruit at index", index, "is", fruits[index])
except IndexError:
    print("Error: Index out of range.")
except ValueError:
    print("Error: Invalid input. Please enter a valid index.")
```
