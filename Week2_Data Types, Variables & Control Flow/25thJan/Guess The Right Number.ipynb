

    # Guess the Right Number Project

Description:
    
1)Asks user to input the upper limit and lower limit to generate the number

2)User is given three trials to guess the generated random number.

```python
import random;

allowed_trials=3
rightGuess=False

#let user input the range for which he wants to guess the number
lowerLimit = int(input("Enter the lower limit to generate a random number "))
upperLimit = int(input("Enter the upper limit to generate a random number "))
if(upperLimit < lowerLimit):
    temp = lowerLimit
    lowerLimit=upperLimit
    upperLimit=temp
#generating random number
random_number = random.randint(lowerLimit, upperLimit)

print("\n The number to be guessed is generated")
print(f"\n You have total {allowed_trials} trials to guess the number ")

while(allowed_trials>0):
    number=int(input("Guess the number: "))
    #checking if number is greater than upper limit
    if(number > upperLimit):
        print("\n You have eneterd the value greater than the defined upper limit.")
        print(f"\n You have {allowed_trials-1} trials left.")
    #checking if number is less than lower limit
    elif(number < lowerLimit):
        print("\n You have eneterd the value less than the defined lower limit.")
        print(f"\n You have {allowed_trials-1} trials left.")
    #checking if number entered is greater than the generated number
    elif(number > random_number and number < upperLimit):
        print("\n The number you have entered is greater than the generated number.")
        print(f"\n You have {allowed_trials-1} trials left.")
    #checking if number entered is less than the generated number
    elif(number < random_number and number > lowerLimit):
        print("\n The number you have entered is less than the generated number.")
        print(f"\n You have {allowed_trials-1} trials left.")
    #checking if entered number is equal to the generated number
    elif(number == random_number):
        print("\n You have guessed the right number")
        #setting rightGuess to true to display result at the end
        rightGuess=True
        break
    
    allowed_trials=allowed_trials-1

if(rightGuess):
    print("\n Congrats...! You are winning a prize.")
else:
    print("\n Sorry... ! You lost the golden chance to win special prize.Correct number was {}".format(random_number))
    
```
