```python
# Functions are defined using the 'def' keyword.

# Example of a simple function:
def greet(name):
    return f"Hello, {name}!"
message = greet("Alice")
print(message) 
```

```python
def calc(a, b):
    add = lambda x, y: x + y
    subtract = lambda x, y: x - y
    multiply = lambda x, y: x * y
    divide = lambda x, y: x / y if y != 0 else "Cannot divide by zero"

    return {
        "Add": add(a, b),
        "Sub": subtract(a, b),
        "Mult": multiply(a, b),
        "Div": divide(a, b)
    }

result = calc(10, 5)
print(result)
```

```python
def myCal(opr,a,b):
    operations = ["+","-","/","*","%"]
    if opr in operations :
        if opr=="+" :
            print(a+b)
        elif opr =="*" :
            print(a*b)
        elif opr == "-" :
            print (a-b)
        elif opr=="/" :
            if b ==0 :
                print("second number cant be zero")
            else : print(a/b)
    else :
        print("Enter valid operation")
```

```python
myCal('+',2,3)
```

```python
def calculator(n1,n2,op):
    if op in ['plus','sum','add','+']:
        print(n1+n2)
    elif op in ['minus','subtract','deduct','-']:
        print(n1-n2)
    elif(op=='*'):
        print(n1*n2)
    elif(op=='/'):
        if(n2==0):
            print("can't divide")
        else:
            print(n1/n2)
    else:
        print("wrong operation")
```

```python
calculator(10,2,'plus')
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operation (+, -, *, /): ")
calculator(num1,num2,op)
```

```python
calc = {
    "+": lambda x, y: x + y,
    "-": lambda x, y: x - y,
    "*": lambda x, y: x * y,
    "/": lambda x, y: x / y if y != 0 else "Error: Division by zero"
}
calc.keys()
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter first number: "))
operator = input("Enter Operator +/-//*: ")
result = calc[operator](num1, num2)
print(f"{num1} {operator} {num2} = {result}")
```

```python
# def with multiple arg 
def add_numbers(a, b):
    return a + b
result = add_numbers(5, 7)
print(result)  # Output: 12
```

```python
# Example of a function with default parameters:
def power(base, exponent=2):
    return base ** exponent
result = power(3)  # Uses default exponent of 2
print(result)  # Output: 9
result = power(3, 3)
print(result)  # Output: 27
```

```python
# Example of a function with variable-length arguments:
def sum_all(*args):
    return sum(args)
total = sum_all(1, 2, 3, 4, 5)
print(total)  # Output: 15
```

```python
# Example of a function with keyword arguments:
def introduce(name, age):
    return f"My name is {name} and I am {age} years old."
intro = introduce(age=30, name="Bob")
print(intro)  
```

```python
# Functions can also be nested, meaning you can define a function inside another function.
# Example of a nested function:
def outer_function(message):
    def inner_function():
        return f"Inner function received: {message}"
    return inner_function() 
result = outer_function("Hello from outer!")
print(result)  # Output: Inner function received: Hello from outer!
```

```python
# Functions can also have annotations that provide hints about the types of parameters and return values.
# Example of a function with annotations:
def add(a: int, b: int) -> int:
    return a + b
result = add(5, 10)
print(result)  # Output: 15
```

```python
# Functions can also be recursive, meaning they can call themselves to solve a problem.
# Example of a recursive function to calculate factorial:
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
result = factorial(5)
print(result)
```

```python
# more for loop examples with pass, break, continue
for i in range(10):
    if i == 5:
        break  # Exit the loop when i is 5
    if i % 2 == 0:
        continue  # Skip even numbers
    print(i)
```

```python
# while loop example
count = 0
while count < 5:
    print(count)
    count += 1  # Increment the counter to avoid infinite loop
    
```

```python
# more while loop example with break, continue, pass
i = 0
while i < 10:
    if i == 5:
        break  # Exit the loop when i is 5
    if i % 2 == 0:
        i += 1
        continue  # Skip even numbers
    if i == 7:
        pass  # Do nothing when i is 7
    print(i)
    i += 1
    
```
